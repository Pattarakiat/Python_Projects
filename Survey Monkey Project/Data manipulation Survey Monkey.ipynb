```python
# Import packages
import pandas as pd
import os
```

```python
# Add current working directry to varriable named pwd and used anconda as a python environment
pwd = os.getcwd()
```

```python
# Import data to variable named dataset and question
edited_import = pd.read_excel(pwd + "\\Data_Survey_Monkey_edited.xlsx", sheet_name="edited_data")
questions_import = pd.read_excel(pwd + "\\Data_Survey_Monkey_edited.xlsx", sheet_name="Questions")
```

```python
# Copy the original data
edited_modified = edited_import.copy()
questions_modified = questions_import.copy()
```

```python
# Take a look at columns name
print(edited_modified.columns)
print(questions_modified.columns)
```

```python
# Drop some columns that we don't use it, firstly we will assign to a variable and create as a list
columns_to_drop = ['Start Date', 'End Date', 'Email Address', 'First Name', 'Last Name', 'Custom Data 1']
columns_to_drop
```

```python
# Drop the columns that we create variable before
edited_drop = edited_modified.drop(columns = columns_to_drop)
edited_drop
```

```python
# Create two lists for melting
edited_id = list(edited_drop.columns)[ :8]
edited_value = list(edited_drop.columns)[8: ]
```

```python
# Change the format of data from wide to long formet by using melt method
edited_melted = edited_drop.melt(id_vars=edited_id, value_vars=edited_value, var_name="Question + Subquestion", value_name="Answer")
edited_melted.head(10)
```

```python
# Requirement1: Find the total number of respondants each questions
# We need to join edited and questions data
# Let's clean question data

# Drop some columns that we don't use it
questions_drop = questions_modified.drop(columns=["raw_question", "raw_sub_question", "sub_question", "Unnamed: 5"])

# remane the columns
questions_drop.rename(columns={"question": "Question", "question+sub_question":"Question + Subquestion"}, inplace=True)
print(questions_drop)
```

```python
# Left join the table by using merge method
dataset_merged = pd.merge(edited_melted, questions_drop, how="left", on="Question + Subquestion")
print("Original data", len(edited_melt))
print("Merged data", len(dataset_merged))
dataset_merged.head(5)
```

```python
# Find the total number of respondants each questions

# subset Answer columns without null values
total_respondents = dataset_merged[dataset_merged["Answer"].notna()]

total_respondents = total_respondents.groupby("Question")["Respondent ID"].nunique().reset_index()
total_respondents.rename(columns={"Respondent ID":"Respondents"}, inplace=True)
total_respondents
```

```python
# Requirement2: Find how many people answered the same answer per questions
same_answer = dataset_merged
same_answer = same_answer.groupby(["Question + Subquestion", "Answer"])["Respondent ID"].nunique().reset_index()
same_answer.rename(columns={"Respondent ID":"Same Answer"}, inplace=True)
same_answer
```

```python
# Join two requirement answers to dataset_mergered
dataset_merged_2 = pd.merge(dataset_merged, total_respondents, how="left", on="Question").merge(same_answer, how="left", on=["Question + Subquestion", "Answer"])
dataset_merged_2["Same Answer"].fillna(0, inplace=True)
print("Original data", len(dataset_merged))
print("Merged data", len(dataset_merged_2))
dataset_merged_2
```

```python
# Rename the columns
output = dataset_merged_2.copy()
output.rename(columns={"Identify which division you work in.-Response":"Division Primary", 
                        "Identify which division you work in.-Other (please specify)":"Division Secondary",
                        "Which of the following best describes your position level?-Response":"Position",
                        "Which generation are you apart of?-Response":"Generation", "Please select the gender in which you identify.-Response":"Gender",
                        "Which duration range best aligns with your tenure at your company?-Response":"Tenure", 
                        "Which of the following best describes your employment type?-Response":"Employee Type"}, inplace=True)


output.head(5)
```

```python
# Export the output as excel
output.to_excel(pwd + "\\Data_output.xlsx", index=False)
```
